```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import Ridge, Lasso
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('D:\Projects_JR\Full-Stack-Projects\AI_class\MLR\ToyotaCorolla - MLR.csv')
data.dropna(inplace=True)
```

```python
# Full features Model
x = data[['Age_08_04', 'KM', 'HP', 'Automatic', 'cc', 'Doors', 'Cylinders', 'Gears']]
y = data['Price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()

model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)

y_pred
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

# Building 3 Models

```python
X1 = data[['Age_08_04', 'KM', 'HP']]
y1 = data['Price']
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=42)
```

```python
print("Model 1 Accuracy:")
print("R2 Score:", r2_score(y1_test, y1_pred))
print("MSE:", mean_squared_error(y1_test, y1_pred))
```

```python
model2 = LinearRegression()
model2.fit(X2_train, y2_train)
y2_pred = model2.predict(X2_test)
```

```python
print("Model 2 Accuracy:")
print("R2 Score:", r2_score(y2_test, y2_pred))
print("MSE:", mean_squared_error(y2_test, y2_pred))
```

```python
X3 = data[['Age_08_04', 'KM', 'HP', 'Automatic', 'cc', 'Doors', 'Cylinders', 'Gears']]
y3 = data['Price']
X3_train, X3_test, y3_train, y3_test = train_test_split(X3, y3, test_size=0.2, random_state=42)
```
